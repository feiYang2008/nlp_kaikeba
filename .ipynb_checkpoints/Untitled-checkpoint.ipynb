```python
line_url = "http://service.shmetro.com/skin/js/pca.js"
line_res = urllib.request.urlopen(line_url)
```

```python
lines_info = line_res.read().decode()
```

```python
lines_info 
```

```python
import re

lines_pat = r"var lines = (\{[\s\S]+?\})"
lines = re.search(lines_pat, lines_info,flags=re.M)
```

```python
lines = lines.group(1)
```

```python
lines = re.sub(r"\r\n|\n","",lines)
```

```python
import json

lines_dict = json.loads(lines)
```

```python
lines_dict
```

```python
len(lines_dict)
```

```python
stations_pat = r"var stations = (\{[\s\S]+?\})"
stations_str = re.search(stations_pat, lines_info).group(1)
```

```python
stations_str = re.sub(r"\r\n|\n","",stations_str)
```

```python
stations_dict = json.loads(stations_str)
```

```python
stations_dict
```

```python
for li,sts in lines_dict.items():
    for st in sts:
        if st not in stations_dict:
            print(st + "not found")
```

```python
station_names = list(set(stations_dict))
```

```python
len(station_names)
```

```python
my_ak = "THBFjRywIrrvLTmGUDje88l3w9Al90pT"
tag = urllib.request.quote("地铁站")
city = urllib.request.quote("上海")
url = 'http://api.map.baidu.com/place/v2/search?query={0}&tag={1}&region={2}&coord_type=1&output=json&ak='+my_ak
```

```python
for station_name in station_names:
    query = urllib.request.quote(station_name + "地铁站")
    request_url = url.format(query,tag,city)
    res = urllib.request.urlopen(request_url)
    
```
